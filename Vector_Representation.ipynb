```python
from matplotlib import pyplot as plt

# 2D vector values
x = [30,20]
y = [10,30]

plt.quiver([0,0],[0,0],x,y,color='b',angles='xy',scale_units='xy',scale=1)
plt.xlim(-10,40)  # Shows the max limit in x direction.
plt.ylim(-10,40)  # Shows the max limit in x direction.
plt.grid()  # Shows the grid in the plot.
plt.show()  # This shows the above plot.


# 3D vector
x = [1,2,3]
y = [2,1,3]
z = [3,2,1]

fig = plt.figure() # This creates an empty canvas as figure.

ax = fig.add_subplot(111, projection='3d') # Creates a 3D axes in plot.

ax.set_xlabel('X-Axis') # This labels the X-axis.
ax.set_ylabel('Y-Axis') # This labels the Y-axis.
ax.set_zlabel('Z-Axis') # This labels the Z-axis.
ax.plot3D(x, y, z, 'red')
plt.show()  # This showstheaboveplot.
```
